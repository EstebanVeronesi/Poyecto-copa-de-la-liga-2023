```python
import pandas as pd
from string import ascii_uppercase as alfabeto
import pickle
```

Extraemos datos de los grupos

```python
todas_tablas = pd.read_html('https://es.wikipedia.org/wiki/Copa_de_la_Liga_Profesional_2023')
```

```python
#Tablas fase de
todas_tablas[7]
todas_tablas[10]
```

```python
for letra, i in zip(alfabeto, range(7,13,3)):
    print(letra, i)
```

```python
dict_tablas = {}
for letra, i in zip(alfabeto, range(7,13,3)):
    df = todas_tablas[i]
    df.rename(columns={df.columns[0]: 'Posicion'}, inplace=True)
    df.rename(columns={df.columns[1]: 'Equipo'}, inplace=True)
    df.rename(columns={df.columns[2]: 'Puntos'}, inplace=True)
    df.rename(columns={df.columns[9]: 'Diferencia'}, inplace=True)
    dict_tablas[f"Grupo {letra}"] = df
print(df)
```

```python
dict_tablas["Grupo B"]
```

```python
with open("dict_table","wb") as output:
    pickle.dump(dict_tablas, output)
```
